```python
import pandas as pd
from datetime import datetime, timedelta

# Read the Traffic Volume data.CSV file as dataframe 

TraffCountCHK3df = pd.read_csv('C:/Users/nsdsc0/OneDrive - WSP O365/DATA_ANALYTICS/Python/SCATS/WEST_20230308.VS.csv')
TraffCountCHK3df


# Convert 'Start_time' column to datetime format
TraffCountCHK3df['Start_time'] = pd.to_datetime(TraffCountCHK3df['Start_time'])

# Create a new column 'Combined_DateTime' by combining 'Date' and 'Start_time'
TraffCountCHK3df['Combined_DateTime'] = TraffCountCHK3df['Date'].astype(str) + ' ' + TraffCountCHK3df['Start_time'].dt.time.astype(str)

# Convert 'Combined_DateTime' column to datetime format
TraffCountCHK3df['Combined_DateTime'] = pd.to_datetime(TraffCountCHK3df['Combined_DateTime'])

# Round 'Combined_DateTime' column down to the nearest 5 minutes
TraffCountCHK3df['Rounded_Combined_DateTime'] = TraffCountCHK3df['Combined_DateTime'].dt.floor('5min')

#Filter the rows in the 'TraffCount3df' DataFrame where the 'Site_ID' column that has the value "2368"
filtered3_df = TraffCountCHK3df[TraffCountCHK3df['Site_ID'] == site_id]

# Pivot the DataFrame to create separate columns for each unique Detector_ID
pivoted3_df = filtered3_df.pivot(index=['Combined_DateTime', 'Site_ID'], columns='Detector_ID', values='Traffic_Volume')

# Reset the index to make the combined DateTime and Site_ID columns regular columns
pivoted3_df = pivoted3_df.reset_index()

# Merge the pivoted DataFrame back into the original DataFrame based on the combined DateTime and Site_ID columns
merged3_df = filtered3_df.merge(pivoted3_df, on=['Combined_DateTime', 'Site_ID'], how='left')

# Remove the 'Combined_DateTime' column
merged3_df.drop('Combined_DateTime', axis=1, inplace=True)

# Rename the 'Rounded_Combined_DateTime' column to 'Combined_DateTime'
merged3_df.rename(columns={'Rounded_Combined_DateTime': 'Combined_DateTime'}, inplace=True)

# Drop the 'Detector_ID' column
merged3_df.drop('Detector_ID', axis=1, inplace=True)

# Remove duplicate values in the 'Combined_DateTime' column
merged3_df.drop_duplicates(subset='Combined_DateTime', inplace=True)

# Drop Cycle_starttime column
merged3_df = merged3_df.drop(['Start_time', 'Traffic_Volume'], axis=1)

# Convert 'Combined_DateTime' column to datetime format
merged3_df['Combined_DateTime'] = pd.to_datetime(merged3_df['Combined_DateTime'])

# Print the updated DataFrame
print(merged3_df)

# Specify the path and filename for the CSV file
merged3 = "C:/Users/nsdsc0/OneDrive - WSP O365/DATA_ANALYTICS/Python/SCATS/merged3.csv"

# Write the DataFrame to a CSV file
merged3_df.to_csv(merged3, index=False)
```
